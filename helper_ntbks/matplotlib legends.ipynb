```python
import matplotlib.pyplot as plt
import pandas as pd
from palmerpenguins import load_penguins
```

```python
data = load_penguins()
```

```python
#remove all NA values
data.dropna(inplace = True)
```

```python
#Create a simple legend in plot

#custom legend- (the legend key is a blue line)
line_handle = plt.Line2D([0, 1], [0, 10], color='blue')
#legend key is a dot
#if linestyle='' is not used, then a point-line gets created. seems like by default, a line key is created by matplotlib
point_handle = plt.Line2D([0],[0], color='red', marker='o', linestyle='')
# Create a simple plot
fig, ax = plt.subplots(figsize = (3,3))
# Add the custom legend
ax.legend([line_handle, point_handle], ['Line', 'dot'], title="Legend")

plt.show()
```

```python
# Define handles with different markers
circle_handle = plt.Line2D([0], [0], color='red', marker='o', markersize=7, linestyle='')
square_handle = plt.Line2D([0], [0], color='blue', marker='s', markersize=7, linestyle='')
triangle_handle = plt.Line2D([0], [0], color='green', marker='^', markersize=7, linestyle='')

# Create a simple plot
fig, ax = plt.subplots(figsize = (5,3))

# Add the custom legend
ax.legend([circle_handle, square_handle, triangle_handle], 
          ['Category 1', 'Category 2', 'Category 3'], 
          title="Categories",
          fontsize = 6,
          title_fontsize = 8)

plt.show()
```
