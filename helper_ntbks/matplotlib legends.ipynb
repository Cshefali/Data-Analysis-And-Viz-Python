#### 1. Legend containing 2 keys: solid line and a dot.

```python
#LEGEND KEY: Blue line. 
line_handle = plt.Line2D([],[], color='blue')
#LEGEND KEY: red dot.
#if linestyle='' is not used, then a point-line gets created. seems like by default, a line key is created by matplotlib
point_handle = plt.Line2D([],[], color='red', marker='o', linestyle='')
```

```python
#Create a simple legend in plot
# Create a simple plot
fig, ax = plt.subplots(figsize = (2,2))
# Add the custom legend- both keys added
ax.legend([line_handle, point_handle], ['Line', 'dot'], title="Legend")

plt.show()
```

### 2. Legend containing different kinds of markers

```python
# Define handles with different markers
circle_handle = plt.Line2D([0], [0], color='red', marker='o', markersize=7, linestyle='')
square_handle = plt.Line2D([0], [0], color='blue', marker='s', markersize=7, linestyle='')
triangle_handle = plt.Line2D([0], [0], color='green', marker='^', markersize=7, linestyle='')

# Create a simple plot
fig, ax = plt.subplots(figsize = (5,3))

# Add the custom legend
ax.legend([circle_handle, square_handle, triangle_handle], 
          ['Category 1', 'Category 2', 'Category 3'], 
          title="Categories",
          fontsize = 6,
          title_fontsize = 8)

plt.show()
```

### 3. Legend containing different linetypes

```python
blue_line = plt.Line2D([], [], color='red', label='species opacity')
#color = 'k' means black color.
black_line = plt.Line2D([], [], color='k', linewidth=1., label='total opacity')
dashed_line = plt.Line2D([], [], color='k', linestyle='--', linewidth=1., label='opacity=1')

#store all the keys in one list.
handles = [blue_line, black_line, dashed_line]
#fetch the labels of all keys
labels = [h.get_label() for h in handles]
```

#### 3a. Vertical legends

```python
#legend keys (vertical)
fig, ax = plt.subplots(1,1, figsize = (3,3))
ax.legend(handles=handles, labels=labels, fontsize='xx-small',loc='upper left')
plt.show()
```

#### 3b. Horizontal legends

```python
fig, ax = plt.subplots(1,1, figsize = (4,2))
ax.legend(handles=handles, labels=labels, fontsize = 'xx-small', loc='upper left', ncol=6)
plt.show()
```
