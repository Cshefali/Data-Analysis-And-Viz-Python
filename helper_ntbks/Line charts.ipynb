```python
import matplotlib.pyplot as plt
import numpy as np
```

### 1. Creating a line between 2 points

The subplots show outputs of variations in plt.plot() function.  
1. Subplot 1- Both x and y coordinates are specified.  
2. Subplot 2- When a marker is specified, points get displayed and not line.  
3. Subplot 3- When only y coordinates are specified, it assumes x coordinates to be 0, 1, 2... so on. 

```python
#x coordinates of 2 points
x = np.array([0, 6, 10])
#y coordinate of the 2 points
y = np.array([5, 15, 23])

#creating layout for 3 subplots in 1 row
fig, ax = plt.subplots(1, 3, constrained_layout = True, figsize = (12,3))

#Subplot 1
ax[0].plot(x, y)
ax[0].set_title('Both parameters x and y specified')

#subplot 2
ax[1].plot(x, y, 'o')
ax[1].set_title('Only show points, not lines')

#subplot 3
ax[2].plot(y)
ax[2].set_title('Only y coordinates specified')
#display figure
plt.show()
```

### 2. Matplotlib Markers

Refer- [Marker table](https://www.w3schools.com/python/matplotlib_markers.asp)
1. Subplot 1- Set shape of the points.  
2. Subplot 2- Usage of shortcut string notation parameter **'fmt'**. Format is **`marker|line shape|color`**.  
3. Subplot 3- 

```python
#x coordinates
x = np.array([1, 2, 3, 4, 5, 6])
#y coordinate
y = x**2

#layout for subplots
fig, ax = plt.subplots(2, 4, figsize = (12, 6))

#subplot 1
ax[0, 0].plot(x, y, 'D')
ax[0, 0].set_title('Set shape of marker', size = 10)

#subplot 2
ax[0, 1].plot(x, y, '*:b')
ax[0, 1].set_title('Fmt 1: dotted blue lines', size = 10)

#subplot 3
ax[0, 2].plot(x, y, 'P--m')
ax[0, 2].set_title('Fmt 2: dashed magenta lines', size = 10)

#subplot 4- ms is markersize
ax[0, 3].plot(x, y, 'P', ms = 10)
ax[0, 3].set_title('Set Marker Size- ms', size = 10)

#suplot 5
ax[1, 0].plot(x, y, '^', ms = 10, mec = 'red')
ax[1, 0].set_title('Set Marker Edge Color- mec', size = 10)

#subplot 6
ax[1, 1].plot(x, y, '^', ms = 10, mec = 'red', mfc = 'yellow')
ax[1, 1].set_title('Set Marker Face Color- mfc', size = 10)

#subplot 7
ax[1, 2].plot(x, y, 'D', ms = 10, mec = 'black', mfc = 'hotpink')
ax[1, 2].set_title('Using all features of plot()', size = 10)

#subplot 8
ax[1,3].plot(x, y, '*-.m')
ax[1,3].set_title('Dashed-dotted magenta line', size = 10)

#display plot
plt.show()
```

### 3. Line Properties

```python
#data
x1 = np.arange(1,8)
y1 = np.sqrt(x)
y11 = x1**2
x2 = x1
y2 = np.log(x2)
```

```python
#subplots layout
fig, ax = plt.subplots(2, 2, constrained_layout = True, figsize = (8,4))

#subplot 1- using full name of arguments
ax[0, 0].plot(x, y, linestyle = "dotted", color = "red", linewidth = 10)
ax[0, 0].set_title('Using full name of arguments', size = 10)

#subplot 2- shortcuts for arguments
ax[0, 1].plot(x, y, ls = "--", c = "green", lw = 4)
ax[0, 1].set_title('Using shortcuts for arguments', size = 10)

#subplot 3- multiple lines using multiple plot() functions
ax[1, 0].plot(x1, y1, c = "red")
ax[1, 0].plot(x2, y2, c = "black")
ax[1, 0].set_title('Multiple plot() functions', size = 10)

#subplot 4- 1 plot function, muliple lines arguments
ax[1, 1].plot(x1, y1, x2, y2)
ax[1, 1].set_title('Multiple (x,y) in same plot()')

#display plot
plt.show()
```

### 3. Titles and Labels

```python
x = np.arange(1,15)
y = np.log(x)

#create dictionary of font properties- color, family and size
font1 = {'color':'blue', 'family':'serif', 'size':10}
font2 = {'color':'red', 'family':'calibri', 'size':15}

#set figure size
plt.figure(figsize = (3,3))
#draw line
plt.plot(x,y, c = "red", ls = "--", lw = 4)
#title
plt.title('Logarithmic line', fontdict = font2, loc = "left")
#xlabel
plt.xlabel('x coordinates', fontdict = font1, loc = "left")
#ylabel
plt.ylabel('y coordinate', fontdict = font1, loc = "bottom")

#display plot
plt.show()
```

### 4. Grid Lines

```python
#layout for subplots
fig, ax = plt.subplots(1, 3, figsize = (8, 3))

#Subplot 1- basic grid lines
ax[0].plot(x, y, '*-.m')
ax[0].set_title('Basic grid lines')
ax[0].grid(color = "green", ls = "--", lw = .5)

#Subplot 2- only horizontal grid lines
ax[1].plot(x, y, '*-.m')
ax[1].set_title('Horizontal grids only')
ax[1].grid(color = "green", ls = "--", lw = .5, axis = 'y')

#Subplot 3- only vertical grid lines
ax[2].plot(x, y, '*-.m')
ax[2].set_title('Vertical grids only')
ax[2].grid(color = "green", ls = "--", lw = .5, axis = 'x')

#display plot
plt.show()
```
