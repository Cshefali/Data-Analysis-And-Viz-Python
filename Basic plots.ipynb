This notebook contains the following charts created using matplotlib and seaborn.  
1. Line Charts
2. Bar Charts- stacked, dodged  
3. Histograms
4. Heatmaps

Random data points have been taken for sake of representation.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Line charts using matplotlib

```python
#data for line chart
years = [2017, 2018, 2019, 2020, 2021, 2022]
orange_yeild = [0.895, 0.91, 0.931, 0.94, 0.92, 0.945]

#plot
#set size of figure
plt.figure(figsize = (4,3))
#create line
plt.plot(years, orange_yeild, c = "red")
#xlabel
plt.xlabel("Year")
#ylabel
plt.ylabel("Yield (tonnes per hectare)")
#title
plt.title("Trend in Orange yield over 6 years")
```

### Trend line for multiple datasets in same graph

```python
#data prep
years = range(2010, 2020)
apple_yield = [0.81, 0.83, 0.87, 0.899, 0.92, 0.93, 0.95, 0.93, 0.88, 0.98]
orange_yield = [0.62, 0.71, 0.55, 0.76, 0.798, 0.799, 0.83, 0.89, 0.93, 0.999]
```

```python
#plot with multiple lines

#set size of figure
plt.figure(figsize = (4,3))
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
plt.plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
plt.plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")
#plot title
plt.title("Fruits yield between 2010-2020")
#xlabel
plt.xlabel("Year")
#ylabel
plt.ylabel("yield (tonne per hectare)")

#show the legend
plt.legend()

#display plot
plt.show()
```

### Line charts using seaborn [TITLES AND SUBPLOTS MISMATCH, ALSO HOW TO GIVE COMMON TITLE AND LEGEND, FIND OUT!!!]

Seaborn has 5 built-in themes: **`darkgrid`, `dark`, `whitegrid`, `white`, `ticks`.**  
The default theme among these is `darkgrid`.

```python
#apply the seaborn style globally

#STYLE- DARKGRID

#set size of whole figure
plt.figure(figsize = (10,8))
#plot title
#plt.title("Fruits yield between 2010-2020")

plt.subplot(2,3,1) #2 rows, 3 columns, subplot index 1

sns.set_style("darkgrid")
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
plt.plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
plt.plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")

#plot title
plt.title("Theme- Darkgrid")
#xlabel
plt.xlabel("Year")
#ylabel
plt.ylabel("yield (tons per hectare)")

#STYLE - WHITEGRID
plt.subplot(2,3,2) #2 rows, 3 columns, subplot index 1

sns.set_style("whitegrid")
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
plt.plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
plt.plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")
#plot title
plt.title("Theme- Whitegrid")
#xlabel
plt.xlabel("Year")
#ylabel
plt.ylabel("yield (tons per hectare)")

#STYLE- TICKS

plt.subplot(2,3,3) #2 rows, 3 columns, subplot index 1

sns.set_style("ticks")
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
plt.plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
plt.plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")
#plot title
plt.title("Theme- Ticks")
#xlabel
plt.xlabel("Year")
#ylabel
plt.ylabel("yield (tons per hectare)")

#STYLE- DARK

plt.subplot(2,3,4) #2 rows, 3 columns, subplot index 1

sns.set_style("dark")
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
plt.plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
plt.plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")
#plot title
plt.title("Theme- Dark")
#xlabel
plt.xlabel("Year")
#ylabel
plt.ylabel("yield (tons per hectare)")

#STYLE- WHITE

plt.subplot(2,3,5) #2 rows, 3 columns, subplot index 1

sns.set_style("white")
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
plt.plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
plt.plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")
#plot title
plt.title("Theme- White")
#xlabel
plt.xlabel("Year")
#ylabel
plt.ylabel("yield (tons per hectare)")

#space between subplots
plt.tight_layout()

#show legend
plt.legend()

#display plot
plt.show()
```
