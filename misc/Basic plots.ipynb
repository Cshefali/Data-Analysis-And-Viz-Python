```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Line charts using matplotlib

```python
#data for line chart
years = [2017, 2018, 2019, 2020, 2021, 2022]
orange_yeild = [0.895, 0.91, 0.931, 0.94, 0.92, 0.945]

#plot
#set size of figure
plt.figure(figsize = (4,3))
#create line
plt.plot(years, orange_yeild, c = "red")
#xlabel
plt.xlabel("Year")
#ylabel
plt.ylabel("Yield (tonnes per hectare)")
#title
plt.title("Trend in Orange yield over 6 years")
```

### Trend line for multiple datasets in same graph

```python
#data prep
years = range(2010, 2020)
apple_yield = [0.81, 0.83, 0.87, 0.899, 0.92, 0.93, 0.95, 0.93, 0.88, 0.98]
orange_yield = [0.62, 0.71, 0.55, 0.76, 0.798, 0.799, 0.83, 0.89, 0.93, 0.999]
```

```python
#plot with multiple lines

#set size of figure
plt.figure(figsize = (4,3))
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
plt.plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
plt.plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")
#plot title
plt.title("Fruits yield between 2010-2020")
#xlabel
plt.xlabel("Year")
#ylabel
plt.ylabel("yield (tonne per hectare)")

#show the legend
plt.legend()

#display plot
plt.show()
```

**Note-** 
1. X and Y labels have been commmented out from subplots `ax[0,1], ax[0,2], ax[1,1]`. This has been done with the intention of keeping a common ylabel for all subplots. This ylabel of left-hand-side plots `ax[0,0] & ax[1,0]` gets shared by all subplots in their respective rows. Same idea applied to xlabels. `ax.set_xlabel("year")` present only for subplots at bottom. They get shared with all subplots in respective columns.  

2. Adding a common legend.  
    - the **`get_legend_handles_labels()` function returns handle names and labels for each component of legend.  
    - For e.g. here, this function will return "apple yield" and "orange yield" as labels and line-symbol (red with diamond, orange with hexagon) as handles. When plt.legend() is used, this function gets automatically called and returns handles and labels which is then used to create legend.  
    - Over here, I have explicitly called this function to store the labels and handles. Then assigned these values to `fig.legend()` at bottom to get a common legend for entire figure.  
    
    - For details, refer: [Creating single legend](https://www.geeksforgeeks.org/how-to-create-a-single-legend-for-all-subplots-in-matplotlib/)

### Line charts using seaborn 

Seaborn has 5 built-in themes: **`darkgrid`, `dark`, `whitegrid`, `white`, `ticks`.**  
The default theme among these is `darkgrid`.  

**The objective was to create 5 subplots in 1 figure, each with a different seaborn style-theme. Seems like it is not possible to create subplots with different themes in 1 figure itself using `.subplots()` function.  
In the subsequent code, I have tried creating such figure using loop and `.subplot()` function. Note that missing 's' in this function. `.subplots()` and `.subplot()` are two different functions.**  

For more details, check- [Seaborn subplots with different styles](https://stackoverflow.com/questions/29235206/changing-seaborn-style-in-subplots)

```python
#apply the seaborn style globally

#add figsize argument to set size of whole figure
fig, ax = plt.subplots(2,3, figsize = (10,8)) #2 rows, 3 columns

#deleting the last panel being created. Here, only 5 subplots are being plotted.
fig.delaxes(ax[1,2])

#plot title
fig.suptitle("Fruits yield between 2010-2020")

#STYLE- DARKGRID

sns.set_style("darkgrid")
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
ax[0,0].plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
ax[0,0].plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")

#subplot title
ax[0,0].set_title("Theme- Darkgrid")
#xlabel
ax[0,0].set_xlabel("Year")
#ylabel
ax[0,0].set_ylabel("yield (tons per hectare)")

#STYLE - WHITEGRID

sns.set_style("whitegrid")
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
ax[0,1].plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
ax[0,1].plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")
#plot title
ax[0,1].set_title("Theme- Whitegrid")
#xlabel
ax[0,1].set_xlabel("Year")
#ylabel
#ax[0,1].set_ylabel("yield (tons per hectare)")

#STYLE- TICKS

sns.set_style("ticks")
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
ax[0,2].plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
ax[0,2].plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")
#plot title
ax[0,2].set_title("Theme- Ticks")
#xlabel
ax[0,2].set_xlabel("Year")
#ylabel
ax[0,2].set_ylabel("yield (tons per hectare)")

#STYLE- DARK

sns.set_style("dark")
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
ax[1,0].plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
ax[1,0].plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")
#plot title
ax[1,0].set_title("Theme- Dark")
#xlabel
ax[1,0].set_xlabel("Year")
#ylabel
ax[1,0].set_ylabel("yield (tons per hectare)")

#STYLE- WHITE

sns.set_style("white")
#line for apples- adding marker arguement highlights each data point on the line
#marker = 'd' is for thin diamond shape
ax[1,1].plot(years, apple_yield, c = "red", label = "apple yield", marker = "d")
#line for oranges
#marker = 'h' is for small hexagon shape
ax[1,1].plot(years, orange_yield, c = "orange", label = "orange yield", marker = "h")
#plot title
ax[1,1].set_title("Theme- White")
#xlabel
ax[1,1].set_xlabel("Year")
#ylabel
ax[1,1].set_ylabel("yield (tons per hectare)")

#saving labels and handles explicitly for common figure legend.
Line, Label = ax[0,0].get_legend_handles_labels()

#adding common legend for the figure
fig.legend(handles = Line, labels = Label, loc = "lower right")

#space between subplots
fig.tight_layout()

plt.show()
```
